```python
import saspy
import os
import pandas as pd
import numpy as np
from IPython.core.display import display, HTML
```

```python
# Provide the path to your configuration file
sas = saspy.SASsession(cfgfile='saspy_config.py')
```

Upload the data from step 1 to SAS OnDemand

```python
sas_code = """
libname scmodel '/home/zheyujiang0/SC';
proc import file = "/home/zheyujiang0/SC/acs_for_model_20240607_2907.csv" out = acs replace;
run;

%inc "/home/zheyujiang0/SC/PROC FORMAT.sas";
data scmodel.acs_20240607;
    set acs;
    %inc "/home/zheyujiang0/SC/FORMAT.sas";
run;

ods exclude all;
ods output oneway = freq;
proc surveyfreq data = scmodel.acs_20240607 (drop = sc_age_cont);
table sc_:;
weight PWGTP;
run;
ods exclude none;

data freq2;
	set freq;
	keep variable category category_n frequency WgtFreq Percent;
    
    array cat_array{*} $ F_sc_:;
    
    do i = 1 to dim(cat_array);
        if not missing(cat_array[i]) then do;
            category = cat_array[i];
        end;
    end;
    
    category_n = coalesce(of sc_:);
	variable = scan(Table, 2, " ");
	if _SkipLine ne "1";
run;


proc sort data = freq2;
	by variable category_n;
run;

title "ACS Weighted Frequency";
ods excel file = "/home/zheyujiang0/SC/acs_freq.xlsx";
proc report data = freq2;
    column 	variable	category Frequency	WgtFreq	Percent;
    define variable /display "Variable";
    define category /display "Category";
    define Frequency /display "Raw Frequency" format = comma10.;
    define WgtFreq/display "Population" f = comma10.;
    define percent/display "Percent" f = 6.2;
run;
ods excel close;

"""

result = sas.submit(sas_code)
# print(result['LOG'])
```
